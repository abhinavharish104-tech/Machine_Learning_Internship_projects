```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df
```

```python
# Set custom theme
sns.set_theme(style="darkgrid")
```

```python
#Distribution plot
plt.figure(figsize=(7,5))
sns.histplot(df['age'], kde=True, color='blue')
plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(["Age Density"])
plt.savefig("age_distribution.png")
plt.show()
```

```python
#Dist plot(Fare)
plt.figure(figsize=(7,5))
sns.histplot(df['fare'], kde=True, color='green')
plt.title("Distribution of Ticket Fare")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.legend(["Fare Density"])
plt.savefig("fare_distribution.png")
plt.show()
```

```python
#Rel plot
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='fare', data=df, hue='survived', palette='coolwarm')
plt.title("Age vs Fare Relationship")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.savefig("age_vs_fare.png")
plt.show()
```

```python
#Rel plot2
plt.figure(figsize=(7,5))
sns.regplot(x='age', y='fare', data=df, scatter_kws={'alpha':0.5})
plt.title("Regression: Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.savefig("regression_age_fare.png")
plt.show()
```

```python
#cat plot
plt.figure(figsize=(7,5))
sns.countplot(x='sex', hue='survived', data=df, palette='Set2')
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived")
plt.savefig("gender_survival.png")
plt.show()
```

```python
#cat plot2
plt.figure(figsize=(7,5))
sns.barplot(x='class', y='survived', data=df, palette='viridis')
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Class")
plt.ylabel("Survival Rate")
plt.savefig("class_survival_rate.png")
plt.show()
```
