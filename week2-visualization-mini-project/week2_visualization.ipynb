```python
import pandas as pd

df = pd.read_csv("Amazon Sale Report.csv")
print(df.head())
print("\nColumns:\n", df.columns)
```

```python
import matplotlib.pyplot as plt
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
# Drop rows where Amount is missing
df = df.dropna(subset=['Amount'])
df
```

```python
daily_sales = df.groupby('Date')['Amount'].sum()

plt.figure(figsize=(8,5))
daily_sales.plot(kind='line', marker='o')
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("line_daily_sales.png")
plt.show()
```

```python
category_sales = df.groupby('Category')['Amount'].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar')
plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Sales Amount")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("bar_category_sales.png")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Qty'], df['Amount'])
plt.title("Quantity vs Sales Amount")
plt.xlabel("Quantity")
plt.ylabel("Amount")
plt.tight_layout()
plt.savefig("scatter_qty_amount.png")
plt.show()
```

```python
status_counts = df['Status'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%')
plt.title("Order Status Distribution")
plt.tight_layout()

plt.savefig("pie_order_status.png")
plt.show()
```
